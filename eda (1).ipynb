```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set a clean figure size
plt.figure(figsize=(8, 6))

# 1. Base boxplot (switching to 'iris' with lowercase and underscores)
sns.boxplot(x="species", y="petal_length", data=iris)

# 2. Overlay stripplot safely 
sns.stripplot(x="species", y="petal_length", data=iris, jitter=True, color="orange", size=5)

# Add custom titles
plt.title('Petal Length Distribution by Species')
plt.xlabel('Species')
plt.ylabel('Petal Length (cm)')

# Display the clean plot
plt.show()
```

```python
# Create the figure with a slightly wider layout to accommodate the legend
plt.figure(figsize=(10, 6))

# Use the 'iris' DataFrame columns
species = iris['species'].unique()
colors = ['red', 'green', 'blue']

# Scatter plot for each species
for i in range(len(species)):
    x_data = iris[iris['species'] == species[i]]
    plt.scatter(
        x_data['sepal_length'], 
        x_data['sepal_width'], 
        c=colors[i], 
        label=species[i].capitalize(),
        s=60,          # Marker size
        edgecolor='w', # White border around dots
        alpha=0.8      # Slight transparency
    )

# Add labels and title
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Sepal Length vs Width by Species', pad=15)

# Place the legend cleanly outside the top-right corner
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)

# Show plot safely
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
```

```python
# From dictionary
data = {'Name': ['Alice', 'Bob'], 'Age': [25, 30]}
df = pd.DataFrame(data)
print (df)
```

```python
# From list of lists
lst = [['Alice', 25], ['Bob', 30]]
df = pd.DataFrame(lst, columns=['Name', 'Age'])
print (df)
```

```python
print (df.head())  	# First 5 rows
print (df.tail())  	# Last 5 rows
print (df.info())  	# Info about DataFrame
print (df.describe())  # Summary statistics
```

```python
# Column selection
print (df['Name'])
print(df[['Name', 'Age']])
```

```python
# Row selection by label and position
print (df.loc[0])
print (df.iloc[1])
```

```python
df[df['Age'] > 25]  # Rows where Age > 25
```

```python
df['Gender'] = ['F', 'M']  # Add new column
print(df)
df.drop('Age', axis=1, inplace=True)  # Remove column
print(df)
```

```python
print(df.isnull())
print(df.dropna())
print(df.fillna(0))
```

```python
df['Age'] = [25,30]  # Add new column
df.groupby('Gender')['Age'].mean()
```

```python
# Merge two DataFrames
df1 = pd.DataFrame({
    'key': [1, 2, 3],
    'name': ['Alice', 'Bob', 'Charlie']
})
df2 = pd.DataFrame({
    'key': [1, 2, 4],
    'age': [25, 30, 40]
})
result=pd.merge(df1, df2, on='key')
print(result)
```

```python
# Join by index
df1 = pd.DataFrame({
    'name': ['Alice', 'Bob', 'Charlie']
}, index=[0, 1, 2])
df2 = pd.DataFrame({
    'age': [25, 30, 35]
}, index=[0, 1, 2])
result=df1.join(df2)
print(result)
```
